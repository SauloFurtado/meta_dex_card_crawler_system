```python
import requests
from pprint import pprint
```

```python
url = 'https://api.pokemontcg.io/v2/cards?q=name:terapagos'

# url = 'https://api.pokemontcg.io/v2/cards?q=set.series:Scarlet&Violet'

# url = 'https://api.pokemontcg.io/v2/cards?q=set.series:Sword&Shield'

# url = 'https://api.pokemontcg.io/v2/cards'

# url = 'https://api.pokemontcg.io/v2/cards?q=id:sv8pt5-163'

# url = 'https://api.pokemontcg.io/v2/sets?qseries:Scarlet&Violet'


headers = {
    'Authorization': 'Bearer seu_token_aqui',
    'Content-Type': 'application/json'
}


try:
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        print(f'Sucesso. Status Code: {response.status_code}')
    else:
        print(f'Erro {response.status_code}: {response.text}')

except requests.exceptions.RequestException as exc:
    print('Error in API', exc)

response.json()
```

```python
resp = response.json()

for card in resp['data']:
    pprint(card)
    break

```
